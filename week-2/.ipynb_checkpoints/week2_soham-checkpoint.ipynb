```python
Q1 = df['Estimated_Deliveries'].quantile(0.25)
Q3 = df['Estimated_Deliveries'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Estimated_Deliveries'] < lower_bound) | (df['Estimated_Deliveries'] > upper_bound)]
print(f"Statistical Outliers Found in Deliveries (Outside {lower_bound:.1f} to {upper_bound:.1f})")
print(outliers[['Year', 'Month', 'Model', 'Estimated_Deliveries']].head())
```

```python
#Time Series Sorting and Feature Generation
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(Day=1))
df = df.sort_values('Date').reset_index(drop=True)

df['Range_per_kWh'] = df['Range_km'] / df['Battery_Capacity_kWh']
df['Delivery_to_Production_Ratio'] = df['Estimated_Deliveries'] / df['Production_Units']

#Lag features and rolling statistics
df['Deliveries_Lag_1'] = df['Estimated_Deliveries'].shift(1)
df['Deliveries_Lag_2'] = df['Estimated_Deliveries'].shift(2)
df['Deliveries_Rolling_Mean_3'] = df['Estimated_Deliveries'].rolling(window=3).mean()

df = df.dropna().reset_index(drop=True)
```

```python
#Categorical Variable Encoding
df_encoded = pd.get_dummies(df, columns=['Region', 'Model'], drop_first=True)

print(df_encoded.head())
```

```python
#Split Data into Train and Test Sets
from sklearn.model_selection import train_test_split

X = df_encoded.drop(columns=['Avg_Price_USD', 'Year', 'Month', 'Source_Type', 'Date'])
y = df_encoded['Avg_Price_USD']

split_index = int(len(X) * 0.8)

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

print("Training records:", len(X_train))
print("Testing records:", len(X_test))
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
#Model Comparison
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression (L2)": Ridge(alpha=1.0),
    "Lasso Regression (L1)": Lasso(alpha=1.0)
}

results = []

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    predictions = model.predict(X_test_scaled)
    
    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    r2 = r2_score(y_test, predictions)
    
    results.append({"Model": name, "MAE": mae, "RMSE": rmse, "R2 Score": r2})

results_df = pd.DataFrame(results)
print(results_df)
```

```python
#GridSearchCV Tuning with Log-Transformed Target
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
import numpy as np

# Apply natural log transformation to stabilize target variance
y_train_log = np.log1p(y_train)

tscv = TimeSeriesSplit(n_splits=5)

# Keep the alpha search space targeted
param_grid = {'alpha': [0.01, 0.1, 1.0, 10.0, 50.0]}

grid_search = GridSearchCV(estimator=Ridge(), 
                           param_grid=param_grid, 
                           cv=tscv, 
                           scoring='r2',
                           n_jobs=-1)

# Fit using the log-transformed target variable
grid_search.fit(X_train_scaled, y_train_log)

print("Best Alpha Parameter:", grid_search.best_params_)
print("Best Log-Transformed Cross-Validation R2 Score:", grid_search.best_score_)

best_model = grid_search.best_estimator_
```

```python
#SARIMAX Modeling and Predictions
from statsmodels.tsa.statespace.sarimax import SARIMAX

ts_data = df.groupby('Date')['Estimated_Deliveries'].sum().sort_index()
ts_data = ts_data.asfreq('MS', fill_value=0)

ts_model = SARIMAX(ts_data, order=(1, 1, 1))
ts_result = ts_model.fit(disp=False)

forecast = ts_result.get_forecast(steps=3)
print(forecast.predicted_mean)
```
